```python
# from langchain.chat_models import ChatOpenAI
# from langchain.agents import initialize_agent, AgentType
# from langchain.tools import Tool
# from langchain.prompts import PromptTemplate
# from langchain.chains import LLMChain

# # Create multiple agents with different roles
# researcher = ChatOpenAI(temperature=0.7, model="gpt-4o-mini")
# writer = ChatOpenAI(temperature=0.5, model="gpt-4o")

# # Define tools for agents
# def research_tool(query):
#     return researcher.predict(f"Research thoroughly: {query}")

# def writing_tool(input):
#     return writer.predict(f"Refine and polish: {input}")

# description1 = """
#  As a histopathology professional with over 30 years of experience and expertise in diagnosing pathology cases, you have extensive knowledge from reputable sources such as:

# 1. College of American Pathologists (CAP) [CAP](https://www.cap.org/member-resources/pathology-careers/pathology-subspecialties/molecular-pathology)
# 2. Diagnostic Histopathology Journal [Diagnostic Histopathology](https://www.diagnostichistopathology.co.uk/)
# 3. Association for Molecular Pathology (AMP) [AMP](https://www.amp.org/)
# 4. Histopathology Journal (Wiley Online Library) [Histopathology Journal](https://onlinelibrary.wiley.com/journal/13652559)
# 5. Advances in Molecular Pathology (ScienceDirect) [Advances in Molecular Pathology](https://www.sciencedirect.com/journal/advances-in-molecular-pathology)
# 6. Journal of Molecular Pathology (MDPI) [Journal of Molecular Pathology](https://www.mdpi.com/topics?journal=jmp&category=1)
# 7. WebPath: The Internet Pathology Laboratory [WebPath](https://webpath.med.utah.edu/ORGAN.html)
# 8. Osmosis [Osmosis](https://www.osmosis.org/library/md/foundational-sciences/anatomy)
# 9. [Advances in Molecular Pathology (ScienceDirect)](https://www.sciencedirect.com/journal/advances-in-molecular-pathology)
# This annual review publication covers current practices and latest developments in molecular pathology. It's a comprehensive source for staying updated on the field.
# 10. [ScienceDirect Topics: Molecular Pathology and Histopathology](https://www.sciencedirect.com/browse/journals-and-books?contentType=JL&contentType=BK&contentType=TB&contentType=HB&contentType=RW&contentType=BS&accessType=openAccess)
# These overview pages provide concise, expert-written summaries on molecular pathology and histopathology, offering quick access to foundational knowledge.

# Your task is to analyze clinical and microscopic features to determine likely diagnoses, suggest special or immunohistochemical stains for confirmation, and provide a well-organized histopathology report.

# When generating a pathology report, adhere to the following comprehensive guidelines to ensure a professional, accurate, and clinically relevant pathology report.

# """

# # Create agents with specific tools
# research_agent = initialize_agent(
#     tools=[Tool(name="Research", func=research_tool, description=description1)],
#     llm=researcher,
#     agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION
# )

# description2 = """
# When generating a pathology report, adhere to the following comprehensive guidelines to ensure a professional, accurate, and clinically relevant pathology report.

# **Report Structure:**

# 1. **Patient Name (in Arabic)**
# 2. **Patient Age**
# 3. **Date Received**
# 4. **Referring Doctor**

# ### Followed by the main sections:

# - **Clinical Data**: Summarize the patient's medical history, presenting symptoms, and any pertinent laboratory or imaging findings.
# - **Gross Examination**: Provide a detailed description of the specimen's appearance, sizes, and any notable gross pathological features, ensuring all components are measured & described.
# - **Microscopic Features**: Provide a thorough analysis of the histopathological findings, including descriptions of tissue architecture, cellular details, and essential criteria for grading and staging the disease, as well as any secondary pathological changes or complications.
# - **Diagnosis**: Clearly articulate the primary diagnosis along with any secondary findings, ensuring consistency with the previously mentioned information and microscopic features. Accurately provide grading, staging, and any relevant disease subtypes and variants.
# - **Comments**: Discuss briefly the rationale behind the diagnosis, addressing the relevant differential diagnoses, then the clinicopathological correlations, and the potential limitations of the current specimen type.
# - **Recommendations**: Offer clear, actionable recommendations for further investigations, particularly IHC stains, to validate the diagnosis and exclude relevant differential diagnoses, and especially stains with prognostic or therapeutic implications.

# Throughout the report, adhere to the latest College of American Pathologists (CAP) guidelines for histopathological reporting. If any critical information is missing, highlight this clearly in the relevant section as [not provided]. 

# For tumor grading and staging, refer to the latest World Health Organization (WHO) and CAP guidelines for the most current analysis.

# Before submission, revise the entire report for spelling, grammar, and overall coherence. Ensure that the language maintains a professional tone while demonstrating high expertise and clinical relevance.

# Adhering to these guidelines will result in a high-quality pathology report that clearly communicates essential findings and supports clinical decision-making.
# """
# writing_agent = initialize_agent(
#     tools=[Tool(name="Write", func=writing_tool, description=description2)],
#     llm=writer,
#     agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION
# )

# # Collaborative workflow
# def multi_agent_workflow(initial_query):
#     # Research phase
#     research_result = research_agent.run(initial_query)
    
#     # Writing phase
#     final_output = writing_agent.run(research_result)
    
#     return final_output

# string = """
# The patient is a 63-year-old female who presented with an elevated dermal mass measuring approximately 3 x 3 cm, located on the back of the
# lower thigh. The lesion has been present for two months and exhibits an indurated base with an ulcerated skin surface. Clinical features were
# suspicious for malignancy, prompting an incisional biopsy of the skin and underlying dermal lesion.

# The specimen consists of a wedge-shaped piece of skin and underlying tissue measuring
# 1 cm in diameter and 2 cm in length. The specimen has been bisected and entirely
# submitted for histological examination.

# Histological examination reveals a dermal spindle cell proliferation extending
# from the superficial dermis deep into the subcutaneous tissue. The lesion is
# composed of nodular fascicles of spindle and stellate-shaped cells arranged in a
# storiform pattern characteristic of dermatofibrosarcoma protuberans (DFSP).
# The neoplastic cells demonstrate moderate to marked atypia, with many cells
# exhibiting nuclear hyperchromasia, pleomorphism, and bizarre morphology. The
# nuclei are elongated and wavy, with some showing prominent nucleoli. Mitotic
# figures are present, including atypical forms, indicating a high proliferative index.
# The tumor infiltrates the underlying subcutaneous fat in a distinctive honeycomb
# or "chicken wire" pattern, where adipocytes are individually entrapped by the
# neoplastic cells. This infiltrative growth pattern is a hallmark feature of DFSP.
# Interspersed within the tumor are collagen bundles, which are often seen in DFSP
# and contribute to its fibrosarcomatous appearance. The overlying epidermis
# shows ulceration, consistent with the clinical presentation.
# No definitive areas of conventional fibrosarcomatous transformation are
# identified in the examined sections, which would be characterized by a
# herringbone pattern and increased cellularity.
# """


# # Example usage
# result = multi_agent_workflow(string)
# print(result)



import os
from langchain.chat_models import ChatOpenAI
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory

class PathologyReportWorkflow:
    def __init__(self, api_key=None):
        # API Key management
        if api_key:
            os.environ["OPENAI_API_KEY"] = api_key
        
        # Initialize memory for context retention
        self.memory = ConversationBufferMemory(memory_key="chat_history")
        
        # Create specialized LLM agents
        self.researcher = ChatOpenAI(
            temperature=0.7, 
            model="gpt-4o-mini", 
            max_tokens=2048
        )
        
        self.writer = ChatOpenAI(
            temperature=0.7, 
            model="gpt-4o", 
            max_tokens=4096
        )
        
        # Research agent description
        research_description = """
     you are a histopathology professional with over 30 years of experience and expertise in diagnosing pathology cases
       

        """
        
        # Writing agent description
        writing_description = """
         you have extensive knowledge from reputable sources such as:
            1. College of American Pathologists (CAP) [CAP](https://www.cap.org/member-resources/pathology-careers/pathology-subspecialties/molecular-pathology)
            2. Diagnostic Histopathology Journal [Diagnostic Histopathology](https://www.diagnostichistopathology.co.uk/)
            3. Association for Molecular Pathology (AMP) [AMP](https://www.amp.org/)
            4. Histopathology Journal (Wiley Online Library) [Histopathology Journal](https://onlinelibrary.wiley.com/journal/13652559)
            5. Advances in Molecular Pathology (ScienceDirect) [Advances in Molecular Pathology](https://www.sciencedirect.com/journal/advances-in-molecular-pathology)
            6. Journal of Molecular Pathology (MDPI) [Journal of Molecular Pathology](https://www.mdpi.com/topics?journal=jmp&category=1)
            7. WebPath: The Internet Pathology Laboratory [WebPath](https://webpath.med.utah.edu/ORGAN.html)
            8. Osmosis [Osmosis](https://www.osmosis.org/library/md/foundational-sciences/anatomy)
            9. [Advances in Molecular Pathology (ScienceDirect)](https://www.sciencedirect.com/journal/advances-in-molecular-pathology)
            This annual review publication covers current practices and latest developments in molecular pathology. It's a comprehensive source for staying updated on the field.
            10. [ScienceDirect Topics: Molecular Pathology and Histopathology](https://www.sciencedirect.com/browse/journals-and-books?contentType=JL&contentType=BK&contentType=TB&contentType=HB&contentType=RW&contentType=BS&accessType=openAccess)
            These overview pages provide concise, expert-written summaries on molecular pathology and histopathology, offering quick access to foundational knowledge.

            Your task is to analyze clinical and microscopic features to determine likely diagnoses, suggest special or immunohistochemical stains for confirmation, and provide a well-organized histopathology report.

        When generating a pathology report, adhere to the following comprehensive guidelines to ensure a professional, accurate, and clinically relevant pathology report.

        **Report Structure:**

        1. **Patient Name (in Arabic)**
        2. **Patient Age**
        3. **Date Received**
        4. **Referring Doctor**

        ### Followed by the main sections:

        - **Clinical Data**: Genrate the patient's medical history, presenting symptoms, and any pertinent laboratory or imaging findings.
        - **Gross Examination**: Provide a detailed description of the specimen's appearance, sizes, and any notable gross pathological features, ensuring all components are measured & described.
        - **Microscopic Features**: Provide a thorough analysis of the histopathological findings, including descriptions of tissue architecture, cellular details, and essential criteria for grading and staging the disease, as well as any secondary pathological changes or complications.
        - **Diagnosis**: Clearly articulate the primary diagnosis along with any secondary findings, ensuring consistency with the previously mentioned information and microscopic features. Accurately provide grading, staging, and any relevant disease subtypes and variants.
        - **Comments**: Discuss briefly the rationale behind the diagnosis, addressing the relevant differential diagnoses, then the clinicopathological correlations, and the potential limitations of the current specimen type.
        - **Recommendations**: Offer clear, actionable recommendations for further investigations, particularly IHC stains, to validate the diagnosis and exclude relevant differential diagnoses, and especially stains with prognostic or therapeutic implications.

        Throughout the report, adhere to the latest College of American Pathologists (CAP) guidelines for histopathological reporting. If any critical information is missing, highlight this clearly in the relevant section as [not provided]. 

        For tumor grading and staging, refer to the latest World Health Organization (WHO) and CAP guidelines for the most current analysis.

        Before submission, revise the entire report for spelling, grammar, and overall coherence. Ensure that the language maintains a professional tone while demonstrating high expertise and clinical relevance.

        Adhering to these guidelines will result in a high-quality pathology report that clearly communicates essential findings and supports clinical decision-making.
        Genrate more details from the prev text for every section.

      -  Clearly define the clinical context of the patient, including the tumor's location, size, and any associated symptoms.
      -  Simplify the scientific language used in the report to make it more accessible for non-specialists or patients.
      -  Include a detailed treatment plan that encompasses advanced options, such as targeted therapies (e.g., Imatinib) for unresectable or recurrent cases.
      -  Add a discussion on potential differential diagnoses and their correlation with the described histopathological features.
      -  Ensure integration of histological findings with clinical and radiological results to provide a comprehensive diagnostic picture.
      -  Provide more detailed follow-up recommendations, specifying required tests and suggested intervals for monitoring
        """
        
        # Research tool
        def research_tool(query):
            return self.researcher.predict(f"with details Analyze pathology case: {query}")
        
        # Writing tool
        def writing_tool(input):
            return self.writer.predict(f"Genrate pathology report: {input}")
        
        # Initialize agents
        self.research_agent = initialize_agent(
            tools=[Tool(name="Research", func=research_tool, description=research_description)],
            llm=self.researcher,
            agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
            memory=self.memory,
            handle_parsing_errors=True,
            verbose=True
        )
        
        self.writing_agent = initialize_agent(
            tools=[Tool(name="Write", func=writing_tool, description=writing_description)],
            llm=self.writer,
            agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
            memory=self.memory,
            handle_parsing_errors=True,
            verbose=True
        )
    
    def generate_report(self, clinical_data):
        try:
            # Research phase
            research_result = self.research_agent.run(clinical_data)
            
            # Writing phase
            final_report = self.writing_agent.run(research_result)
            
            return final_report
        
        except Exception as e:
            return f"Error generating report: {str(e)}"

# Example usage
def main():
    workflow = PathologyReportWorkflow()
    
    clinical_case = """
  The patient is a 63-year-old female who presented with an elevated dermal mass measuring approximately 3 x 3 cm, located on the back of the
lower thigh. The lesion has been present for two months and exhibits an indurated base with an ulcerated skin surface. Clinical features were
suspicious for malignancy, prompting an incisional biopsy of the skin and underlying dermal lesion.


The specimen consists of a wedge-shaped piece of skin and underlying tissue measuring
1 cm in diameter and 2 cm in length. The specimen has been bisected and entirely
submitted for histological examination


Histological examination reveals a dermal spindle cell proliferation extending
from the superficial dermis deep into the subcutaneous tissue. The lesion is
composed of nodular fascicles of spindle and stellate-shaped cells arranged in a
storiform pattern characteristic of dermatofibrosarcoma protuberans (DFSP).
The neoplastic cells demonstrate moderate to marked atypia, with many cells
exhibiting nuclear hyperchromasia, pleomorphism, and bizarre morphology. The
nuclei are elongated and wavy, with some showing prominent nucleoli. Mitotic
figures are present, including atypical forms, indicating a high proliferative index.
The tumor infiltrates the underlying subcutaneous fat in a distinctive honeycomb
or "chicken wire" pattern, where adipocytes are individually entrapped by the
neoplastic cells. This infiltrative growth pattern is a hallmark feature of DFSP.
Interspersed within the tumor are collagen bundles, which are often seen in DFSP
and contribute to its fibrosarcomatous appearance. The overlying epidermis
shows ulceration, consistent with the clinical presentation.
No definitive areas of conventional fibrosarcomatous transformation are
identified in the examined sections, which would be characterized by a
herringbone pattern and increased cellularity.

DERMAL AND SUBCUTANEOUS SPINDLE CELL NEOPLASM WITH FEATURES HIGHLY SUGGESTIVE OF
DERMATOFIBROSARCOMA PROTUBERANS (DFSP).
    """
    
    report = workflow.generate_report(clinical_case)
    print(report)

if __name__ == "__main__":
    main()
```
